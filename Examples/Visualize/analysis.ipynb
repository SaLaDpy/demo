# Python Visulization of SpringSaLaD Outputs

This notebook demonstrates different ways visualize the CSV outputs generate by SpringSaLaD.

```python
#Import necessary functions and data
from saladpy.Visualization import *
from saladpy.data_locator import data_file_finder
from saladpy.display_info import *
from saladpy.Project_reader import Describe_input_file
from saladpy.MolDraw.MolDraw_exec import display_molecules
from saladpy.format_directory import format_directory
```

```python
#Create the relative path to your desired SIMULATIONS or FOLDER directory
search_directory = os.path.join('..', '..', 'Examples', 'GUI_results', 'Nephrin-Nck-NWasp', 'Final_version_test_SIMULATIONS', 'Simulation0_SIM_SIMULATIONS')
#search_directory = os.path.join('..', 'Simulate', 'modified_input_FOLDER')

#if the current path goes to a SIMULATIONS direcotry, convert it to the FOLDER directory one level down
search_directory = format_directory(search_directory)

#Get a description of the input file in that directory
Describe_input_file(search_directory, links=False, reactions=True, kinetics=True)
```

```python
#Display a 2D projection of each moleclue in the input file
display_molecules(search_directory)
```

## Counts of bonds, moleclues, and sites over time

### Options:
- **data_selection**: pick the file that will be used for plotting
    - 'FullBondData.csv' number of each type of bond
    - 'FullCountData.csv' number of each type of moleclue grouped by FREE, BOUND, and TOTAL (default)
    - 'FullStateCountData.csv' number of each site in each state grouped by FREE, BOUND, and TOTAL
- **indicies**: pick the columns that will be graphed over time by passing in a list of integers (it displays all columns by default)
- **list_options**: prints the available columns and their corresponding indicies (default True)
- **legend_right**: displays the legend off to the side of the plot (default True)
- **fill**: displays one standard deviation bounds (default True)

```python
Counter.plot(search_directory, data_selection='FullCountData.csv', indicies=[2,5,8], list_options=True, legend_right=True, fill=True)
```

# Average z position over time

### Options:
- **mode**: pick what type of data will be plotted
    - 'mol' finds the average position for any site that's part of a particular moleclue (default)
    - 'color' finds the average position for any site of a particular color
    - 'site' finds the average position for any site
- **indicies**: pick the columns that will be graphed over time by passing in a list of integers (it displays all columns by default)
- **list_options**: prints the available columns and their corresponding indicies (default True)
- **verbose**: displays both the site name and its corresponding moleclue in the legend for color or site mode (default False)
- **legend_right**: displays the legend off to the side of the plot (default True)
- **fill**: displays one standard deviation bounds (default True)

```python
Average_z_pos.plot(search_directory, mode='mol', indicies=[], list_options=True, verbose=False, legend_right=True, fill=True)
```

## Radius of gyration scatter plot and bonds per molecule histogram

### Options:
- times: list of all the time points that will be used for plotting (picks last available time by default)
- size_threshold: only analyze clusters with more moleclues than the threshold value (default 1)
- bonds_hist: display bonds per moleclue histogram below radius of gyration plot

```python
Spatial_Analysis.plot(search_directory, times=[0.0078, 0.008], size_threshold=2, bonds_hist=True)
```

## Cluster size distribution

### Options:
- times: list of all the time points that will be used for plotting (picks last available time by default)
- bins: list of the edges of the bins (default is no bins)

```python
Histogram.plot(search_directory, times=[0.008, 0.004], bins=[1,2,3])
```

## Cluster composition

### Options:
- times: list of all the time points that will be used for plotting (picks last available time by default)
- special_clusters: list of sizes that controls what clusters will be displayed (all sizes are displayed by default)

```python
Composition.plot(search_directory, times=[0.008, 0.012], special_clusters=[5,6,7])
```

## Bound Fraction

### Options:
- times: list of all the time points that will be used for plotting (picks last available time by default)
- run: specifies the desired run number
- size_threshold_mean: only clusters greater than this size will be included in the mean on the scatter plot
- hist: displays histogram of Saturation Index (bound sites / free sites)

```python
Bound_fraction.plot(search_directory, times=[0.024], run=2, size_threshold_mean=5, hist=True)
```

## Time Courses

### Options:
- data_selection: select the type of data that will be plotted
    - 'rg' plots cluster radius of gyration (default)
    - 'cs' plots cluster size (number of moleclues in a cluster)
    - 'mr' plots maximum cluster radius relative to the center of mass of the cluster
- indicies: list that describes what data gets plotted
    - 0 for maximum (default)
    - 1 for average (default)
- size_threshold: only analyze clusters with more moleclues than the threshold value (default 1)

```python
Spatial_Analysis.time_course(search_directory, data_selection='rg', indicies=[0,1], size_threshold=1)
```
